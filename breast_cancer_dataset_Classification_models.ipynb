# **EDA**
Exploratory Data Analysis (EDA) is a crucial step in the data analysis process that involves investigating datasets to summarize their main characteristics, often using visualizations. It helps in understanding patterns, detecting anomalies, and determining relationships between variables before applying models.
Key aspects of EDA include:
- Data Summarization: Identifying missing values
- Visualizations: Using histograms, scatter plots, box plots, and correlation matrices to interpret data distributions and relationships.
- Feature Engineering: Identifying relevant variables and transforming data for better model performance.

# **KNN**
K-Nearest Neighbors (KNN) is a simple, yet powerful, supervised machine learning algorithm used for classification and regression tasks. It works by finding the K closest data points (neighbors) to a given input and making predictions based on their values.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("/content/breast_cancer_data.csv")

le = LabelEncoder() # converting categorical data into numerical
data['diagnosis'] = le.fit_transform(data['diagnosis'])
data.head()
```

```python
x = data.iloc[:, 2:-1] # extracting x and y variables
y = data.iloc[:, [1]]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # splitting data into training and testing
scaler = StandardScaler() # normalizing the data
X_scaled = scaler.fit_transform(x)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)# calling the KNN classifier with k value as 3
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test) # predicting
# performance metrics for KNN Classifier
print("KNN")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Got an accuracy of 92%, which is the lowest when compared to other models

# **Logistic Regression**

- Type: Linear model used for classification.
- How it Works: It predicts the probability of an instance belonging to a particular class using the sigmoid function, which transforms output values into probabilities.
- Use Cases: Binary classification (e.g., spam detection, medical diagnosis).
- Pros: Simple, interpretable, works well with linearly separable data.
- Cons: Struggles with complex patterns and non-linearly separable data.


```python
lr = LogisticRegression(max_iter=10000, random_state=42) # calling the Logistic Regression
y_pred = lr.predict(X_test)# predicting
# performance metrics for Logistic Regression Classifier
print("Logistic Regression")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Got an accuracy of 95% which is better than KNN

# **Decision Tree**
- Type: Non-linear, tree-based model.
- How it Works: It splits the data into branches based on feature conditions, creating a hierarchy of decisions.
- Use Cases: Fraud detection, customer segmentation.
- Pros: Easy to understand, works with both categorical & numerical data.
- Cons: Prone to overfitting without pruning.

```python
dt = DecisionTreeClassifier(random_state=42) # calling the Decision Tree classifier
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test) # predicting
# performance metrics for Decision Tree Classifier
print("Decision Tree")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Got an accuracy of 94% which is good

# **Random Forest**
- Type: Ensemble of decision trees.
- How it Works: It builds multiple decision trees and combines their outputs for better accuracy.
- Use Cases: Image classification, risk assessment.
- Pros: Robust to overfitting, handles large datasets.
- Cons: Computationally expensive.


```python
rf = RandomForestClassifier(random_state=42) # calling the Random Forest Classifier
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test) # predicting
# performance metrics for Random Forest Classifier
print("Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Got an accuracy of 96% which is better than decision tree

# **SVM**
- Type: Linear or non-linear classifier.
- How it Works: Finds the optimal hyperplane that separates different classes.
- Use Cases: Text classification, face recognition.
- Pros: Works well in high-dimensional space.
- Cons: Slower on large datasets.


```python
svm = SVC(kernel='linear', random_state=42) # calling the SVM classifier
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test) # predicting
# performance metrics for SVM Classifier
print("SVM")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Got an accuracy of 95% which is slightly lower than random forest

# **Naive Bayes**
- Type: Probabilistic model based on Bayes’ theorem.
- How it Works: Assumes features are independent and calculates probabilities.
- Use Cases: Sentiment analysis, spam filtering.
- Pros: Fast, works well with small datasets.
- Cons: Assumes independence among features, which isn’t always true.


```python
gnb = GaussianNB() #calling the Naive Bayes classifier
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
#performance metrics for Naive Bayes Classifier
print("Naive Bayes")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Got an accuracy of 97% which indicates it is well trained and a very good model

# ***Evaluating each model's performance***

```python
models = {
    "Random Forest": RandomForestClassifier(),
    "Naïve Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Logistic Regression": LogisticRegression(),
    "Decision tree": DecisionTreeClassifier()
}
performance = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    performance[name] = [accuracy, precision, recall, f1]
    #performance metrics for all the models
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    print("-" * 60)
```

***Let us visualise our model's performance and conclude which model gave the best performance***

```python
performance_df = pd.DataFrame(performance, index=["Accuracy", "Precision", "Recall", "F1-Score"])
# plotting ou rmodel's performance
plt.figure(figsize=(10, 6))
performance_df.T.plot(kind='bar')
plt.title("Model Performance Comparison")
plt.xlabel("Models")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.legend(loc="lower right")
plt.show()
```



***Although all the models achieved similar performance, we can identify and select Naive Bayes as the most optimal model.***
